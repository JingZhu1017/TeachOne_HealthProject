# Question 1. Construct a simple Multi-Morbidity Index

```python
data = pd.read_excel('LM3_MM_Index.xlsx')
```

```python
print(data)
```

```python
#Cover the diagnosises data into binary formats in order to be used to display the presense of each diagnosis.
data.loc[data['MI'] == 'MI', 'MIDX']=1
data.loc[data['MI'].isnull(), 'MIDX']=0
data.loc[data['CHF'] == 'CHF', 'CHFDX']=1
data.loc[data['CHF'].isnull(), 'CHFDX']=0
data.loc[data['DM'] == 'DM', 'DMDX']=1
data.loc[data['DM'].isnull(), 'DMDX']=0
data.loc[data['AA'] == 'AA', 'AADX']=1
data.loc[data['AA'].isnull(), 'AADX']=0

#View the dataframe head
data.head()
```

```python
#Put data ouput into dataframe
df=pd.DataFrame(data)

#Show the dataframe
df
```

```python
#Calculate data values 
df['losMIDX'] = df['LOS']*df['MIDX']
df['losCHFDX'] = df['LOS']*df['CHFDX']
df['losDMDX'] = df['LOS']*df['DMDX']
df['losAADX'] = df['LOS']*df['AADX']

#Calculate Control values
df['losConMI']=(1-df['MIDX'])*df['LOS']
df['losConCHF']=(1-df['CHFDX'])*df['LOS']
df['losConDM']=(1-df['DMDX'])*df['LOS']
df['losConAA']=(1-df['AADX'])*df['LOS']

#Put data ouput into dataframe
df1 = pd.DataFrame(df)
df1
```

```python
#Calculate the data value
df1['data_MI']=(df1['losMIDX'].sum()/df1['MIDX'].sum()).round(decimals=2)
df1['data_CHF']=(df1['losCHFDX'].sum()/df1['CHFDX'].sum()).round(decimals=2)
df1['data_DM']=(df1['losDMDX'].sum()/df1['DMDX'].sum()).round(decimals=2)
df1['data_AA']=(df1['losAADX'].sum()/df1['AADX'].sum()).round(decimals=2)

#Calculate the controls value
df1['Con_MI']=(df1['losConMI'].sum()/(1-df1['MIDX']).sum()).round(decimals=2)
df1['Con_CHF']=(df1['losConCHF'].sum()/(1-df1['CHFDX']).sum()).round(decimals=2)
df1['Con_DM']=(df1['losConDM'].sum()/(1-df1['DMDX']).sum()).round(decimals=2)
df1['Con_AA']=(df1['losConAA'].sum()/(1-df1['AADX']).sum()).round(decimals=2)

#Put data output into DataFrame
df2=pd.DataFrame(df1)
df2
```

```python
###To calculate the estimate
#Estimate = LOS for cases - LOS Control

#Calculate Estimate Values
df2['Est_MI']=df2['data_MI']-df2['Con_MI']
df2['Est_CHF']=df2['data_CHF']-df2['Con_CHF']
df2['Est_DM']=df2['data_DM']-df2['Con_DM']
df2['Est_AA']=df2['data_AA']-df2['Con_AA']

#Put data output into dataframe
df3 = pd.DataFrame(df2)
df3
```

```python
#Caculate Severity
df3['severity']=((df3['MIDX']*df3['Est_MI'])+(df3['CHFDX']*df3['Est_CHF']
                 +(df3['DMDX']*df3['Est_DM'])+(df3['AADX']*df3['Est_AA'])).round(decimals=2))

#Put data output into dataframe
df4 = pd.DataFrame(df3)
df4
```

# Question 2. Calculate the likelihood ratio associated

```python
data = pd.read_excel('LM3_MM_Index.xlsx')
```

```python
print(data)
```

```python
data['N']=data['N'].astype(float)
print(data)
```

```python
AvgLos = (data["LOS"]*data["N"]).sum()/data["N"].sum()
print(AvgLos)
```

```python
print(data)
```

```python
data.replace(to_replace = ['MI','CHF','DM','AA'],
              value = ['1','1','1','1'],
              inplace = True)
data = data.fillna(0)
data["MI"]=data["MI"].astype(float)
data["CHF"]=data["CHF"].astype(float)
data["DM"]=data["DM"].astype(float)
data["AA"]=data["AA"].astype(float)
```

```python
data['TempMILong'] = np.where(data['LOS']>AvgLos,'1','0')
data["TempMILong"] = data["TempMILong"].astype(float)
data['TempMIShort'] = np.where(data['LOS']<=AvgLos,'1','0')
data["TempMIShort"] = data["TempMIShort"].astype(float)
data
```

```python
data['TempAALong'] = np.where(data['LOS']>AvgLos,'1','0')
data["TempAALong"] = data["TempAALong"].astype(float)
data['TempAAShort'] = np.where(data['LOS']<=AvgLos,'1','0')
data["TempAAShort"] = data["TempAAShort"].astype(float)
data
```

```python
MILong=(data['MI']*data['N']*data["TempMILong"]).sum()
print(MILong)
MIShort=(data['MI']*data['N']*data["TempMIShort"]).sum()
print(MIShort)

AALong=(data['AA']*data['N']*data["TempAALong"]).sum()
print(AALong)
AAShort=(data['AA']*data['N']*data["TempAAShort"]).sum()
print(AAShort)
```

```python
LongLOS=(data['N']*data["TempMILong"]).sum()
print(LongLOS)
ShortLOS=(data['N']*data["TempMIShort"]).sum()
print(ShortLOS)
AALongLOS=(data['N']*data["TempAALong"]).sum()
print(AALongLOS)
AAShortLOS=(data['N']*data["TempAAShort"]).sum()
print(AAShortLOS)
```

```python
MILIkelihoodRatio = [(MILong/LongLOS)/(MIShort/ShortLOS)]
print(MILIkelihoodRatio)
AALIkelihoodRatio = [(AALong/AALongLOS)/(AAShort/AAShortLOS)]
print(AALIkelihoodRatio)
```

```python
##### Finding Log####
data['log']=np.log10(data["LR"])
data.head()
```

```python
#### Finding the sum of log####
result = data.groupby("id")["log"].sum()

###Put result output data into dataframe###
result_log=pd.DataFrame(result)
result_log
```

```python
#### Calculating Odds ####
result_log["Odds"]=np.exp(result_log["log"])
result_Odds=pd.DataFrame(result_log)
result_Odds
```

```python
result_Odds["Probability"]=((result_Odds["Odds"])/(1+(result_Odds["Odds"]))).round(5)
result_Prob=pd.DataFrame(result_Odds)
result_Prob
```
