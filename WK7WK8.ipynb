```python
#Convert data types

samc['Score'] = pd.to_numeric(samc['Score'])
samc['Sample'] = pd.to_numeric(samc['Sample'])
samc['Measure Start Date'] = pd.to_datetime(samc['Measure Start Date'])
samc['Measure End Date'] = pd.to_datetime(samc['Measure End Date'])
samc.dtypes
```

```python
#Calcuate Midway Point: ( [Measure Start Date] + [Measure End Data] ) / 2  

samc['MidP'] = samc['Measure Start Date'] + (samc['Measure End Date'] - samc['Measure Start Date'])/2

#Calcuate Rate of Overuse: [Score] / [Sample]
samc['RateOveruse'] = samc['Score']/samc['Sample']

#Grand Rate: SUM (Score) / SUM (Sample)
samc['Grandrate'] = sum(samc['Score'])/sum(samc['Sample'])

#Upper Limit: Grand Rate + 1.96*SQRT((Grand Rate*(1-Grand Rate))/Sample)
#Lower Limit: Grand Rate - 1.96*SQRT((Grand Rate*(1-Grand Rate))/Sample)

samc['UCL'] = samc['Grandrate'] + (1.96 * np.sqrt((samc['Grandrate'] * (1 - samc['Grandrate']))/samc['Sample']))
samc['LCL'] = samc['Grandrate'] - (1.96 * np.sqrt((samc['Grandrate'] * (1 - samc['Grandrate']))/samc['Sample']))
samc
```

```python
#Create the p-chart
plt.figure(figsize=(15,5))
plt.plot('Measure Start Date', 'RateOveruse', data=samc, marker = 's', markerfacecolor = 'blue', linewidth =1, color = 'blue')
plt.plot('Measure Start Date', 'UCL', data=samc, markersize = 0, color = 'red', linewidth =1)
plt.plot('Measure Start Date', 'LCL', data=samc, markersize = 0, color = 'red', linewidth =1)
plt.rotation=60
plt.legend()
plt.title('Rate of Post-Op Antibiotic Overuse: Apr 2013 - Jan 2015')
plt.xlabel('Measure Start Date')
plt.ylabel('Rate of Overuse')
```

```python
#Create dataframe for beginning to determine Expected Deviation
#Select specific columns from df dataframe
df1=pd.DataFrame(df.iloc[:,0:21])#First 20 columns of data frame(C0 to C20)
df1
```

```python
#Expected Devation
#Create new dataframe using data from df2 and calculate risk
df2=pd.DataFrame()
for (columnName, columnData) in df1.iteritems():
    df2[columnName+'Risk']=(df1[columnName])*(1-df1[columnName])
df2
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Q3Data.xlsx')
data
```

```python
#Check the data types
data.dtypes
```

```python
#renaming columns
df=data.rename(columns={'Risk of Infection for each Patient':'Risk P1',
                        'Unnamed: 2':'Risk P2',
                        'Unnamed: 3':'Risk P3',
                        'Unnamed: 4':'Risk P4',
                        'Unnamed: 5':'Risk P5',
                        'Unnamed: 6':'Risk P6'},)
df
```

```python
#Calculate Number of Cases per week
df['N']=df[['Risk P1','Risk P2','Risk P3','Risk P4','Risk P5','Risk P6']].count(1)
df
```

```python
#Calculate Observed Rate
df['Obs Rate']=df['Number Infected']/df['N']
df
```

```python
### Step2 Calculate Expected Rate
#Expected Rate = Average of each week
df['Exp Rate']=df[['Risk P1','Risk P2','Risk P3','Risk P4','Risk P5','Risk P6']].mean(1)
df
```

```python
### Step3 Calculate Expected Deviation
#Expected Deviation = The square root of the sum of 
#[Rate (of each pt)*1-Risk(of each pt)]in each week/Number of Cases

#Removing nulls, replacing with 0
df1=df.fillna(0)

df['Exp Stdev']=(np.sqrt((df1['Risk P1']*(1-df1['Risk P1']))+
                        (df1['Risk P2']*(1-df1['Risk P2']))+
                        (df1['Risk P3']*(1-df1['Risk P3']))+
                        (df1['Risk P4']*(1-df1['Risk P4']))+
                        (df1['Risk P5']*(1-df1['Risk P5']))+
                        (df1['Risk P6']*(1-df1['Risk P6']))))/df['N']
df
```

```python
#Step4 Get T-Value
#Calculate degrees of freedom
df['DoF']=df['N']-1
df
```

```python
#Insert t-value from table (this can be found online)5dof = 2.5705,
#3dof=3.1824,2dof=4.3027
df['T-value']=[2.5705,2.5705,3.1824,3.1824,2.5705,4.3027]
df
```

```python
#Calculate Control Limits

#Calculating UCL and LCL
df['UCL']=(df['Exp Rate']+(df['T-value']*df['Exp Stdev']))
df['LCL']=(df['Exp Rate']-(df['T-value']*df['Exp Stdev']))

#Replacing any limit less than 0 with 0 and greater than 1 with 1
df.loc[df.LCL<0,'LCL']=0
df.loc[df.UCL>1,'UCL']=1
df
```

```python
#Plot a Risk - Adjust P Chart
plt.plot('Week','Obs Rate',data=df,marker='o',color='blue')
plt.plot('Week','UCL',data=df,markersize=0,color='red')
plt.plot('Week','LCL',data=df,markersize=0,color='red')
plt.legend()
plt.title('Probability of Infection Over Time')
plt.xlabel('Week')
plt.ylabel('Probability of Infection')
```

# Conclusion

No, we are not having more infections than expected from the patients' conditions beacause our observations are within the upper and lower control limits.

```python
#Import data files 
x1=pd.ExcelFile('W7Q8.xlsx')
df=x1.parse('data')
df
```

```python
df=df.rename(columns={"Services - Date of Service":"service_and_date"})
df=(df.set_index(df.columns.drop('service_and_date',1).tolist())
    .service_and_date.str.split(r'\n',expand=True)
    .stack()
    .reset_index()
    .rename(columns={0:'service_and_date'})
    .loc[:,df.columns])
df
```

```python
df[['Service','Service Date']]=df['service_and_date'].str.split(" - ",expand=True)
df=df.drop(columns=['service_and_date'])
df
```

```python
df['year_month']=df['Service Date'].str[:7]
df=df.sort_values(by=['year_month'])
unique_ym=df.year_month.unique()
unique_ym.sort()
unique_ym
```

```python
df['time_period']=df.groupby(['year_month']).ngroup()+1
df
```

```python
df_results=df.groupby(['time_period']).size().reset_index(name='Number_Screened')
df_results
```

```python
df_n = df.groupby('time_period')['Service'].apply(lambda x:x[x == 'Neurology'].count()).reset_index().rename(columns={'Service':'Neuro_ref'})
df_n
```

```python
df_results=pd.merge(df_results,df_n,on='time_period',how='outer')
df_results
```

```python
df_results['prob_ref']=df_results['Neuro_ref']/df_results['Number_Screened']
df_results
```

```python
total_screened=df_results['Number_Screened'].sum()
total_neuro_ref=df_results['Neuro_ref'].sum()
print('Total Screened:'+str(total_screened))
print('Total Neurology Referrals:'+str(total_neuro_ref))

p_bar=total_neuro_ref/total_screened
print('P_Bar:'+str(p_bar))
```

```python
df_results['std_dev']=np.sqrt(p_bar*(1-p_bar)/df_results['Number_Screened'])
df_results['UCL']=p_bar + 1.96 * df_results['std_dev']
df_results['LCL']=p_bar - 1.96 * df_results['std_dev']
df_results
```

```python
df_results['revised_LCL']=np.where(df_results['LCL']<0,0,df_results['LCL'])
df_results
```

```python
#Create the control chart
plt.plot('time_period','prob_ref',data=df_results,marker='o',markerfacecolor='blue',
        markersize=3,color='blue',linewidth=1)
plt.plot('time_period','UCL',data=df_results,marker='',color='red',linewidth =1)
plt.plot('time_period','revised_LCL',data=df_results,marker='',color='red',linewidth =1)
plt.legend(loc='upper right',bbox_to_anchor=(1,1.3))
plt.title('The p- chart for the probability of a subsequent neurological visits in mothers who had an ultrasound screening')
plt.xlabel('Time Period',fontsize=5)
plt.ylabel('The probability of visit to Neurology')
```
